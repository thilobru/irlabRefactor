# Plots for Topic Relevance

### libraries

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

## read evaluation files 

```python
vertical_prototype = pd.read_csv(os.getcwd()+'/evaluations/query_expansion.tsv', sep='\t', header=0)
text_sentiment = pd.read_csv(os.getcwd()+'/evaluations/text_sentiment.tsv', sep='\t', header=0)
refined_prototype = pd.read_csv(os.getcwd()+'/evaluations/refined_prototype.tsv', sep='\t', header=0)
queryANDsentiment = pd.read_csv(os.getcwd()+'/evaluations/query_exp_with_text_sent.tsv', sep='\t', header=0)
```

## make bar plots for average TP's

```python
# calculate TP averages
VP_avg_tp_pro = vertical_prototype["TP Pro"].mean()
VP_avg_tp_con = vertical_prototype["TP Con"].mean()

TS_avg_tp_pro = text_sentiment["TP Pro"].mean()
TS_avg_tp_con = text_sentiment["TP Con"].mean()

RP_avg_tp_pro = refined_prototype["TP Pro"].mean()
RP_avg_tp_con = refined_prototype["TP Con"].mean()

QE_TS_avg_tp_pro = queryANDsentiment["TP Pro"].mean()
QE_TS_avg_tp_con = queryANDsentiment["TP Con"].mean()
```

```python
# normal bar plot
Names = ['VP Pro','VP Con','TS Pro','TS Con','RP Pro', 'RP Con', 'QE&TS Pro', 'QE&TS Con']
Averages = [VP_avg_tp_pro, VP_avg_tp_con, TS_avg_tp_pro, TS_avg_tp_con, RP_avg_tp_pro, RP_avg_tp_con, QE_TS_avg_tp_pro, QE_TS_avg_tp_con]

plt.bar(Names, Averages, color=['green','green','brown','brown','teal','teal', 'orange','orange'])
plt.title('Average True Positives over all Topics', fontsize=14)
# plt.xlabel('Trials')
plt.ylabel('Average TP')
plt.grid(True)
plt.show()
```

```python
# stacked bar plot
Names = ['Vertical Prototype','Text Sentiment','Refined Prototype']
Pros = [VP_avg_tp_pro, TS_avg_tp_pro, RP_avg_tp_pro]
Cons = [VP_avg_tp_con, TS_avg_tp_con, RP_avg_tp_con]
width = 0.5

fig, ax = plt.subplots()

ax.bar(Names, Pros, width, label='Pro', color='teal')
ax.bar(Names, Cons, width, bottom=Pros, label='Con', color='brown')

ax.set_ylabel('Average TP')
ax.set_title('Average True Positives over all Topics', fontsize=14)
ax.legend()
ax.grid(True)

plt.show()
```
